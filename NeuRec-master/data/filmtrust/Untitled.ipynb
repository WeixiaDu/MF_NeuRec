```python
import pandas as pd
```

```python
train_data = pd.read_csv('./ratings.txt',sep = ' ',names=['user_id', 'item_id', 'rating'])
train_data.head()
```

```python
train_dict = {}
for line in train_data.itertuples():
    u = line[1] - 1
    i = line[2] - 1
    r = line[3]
    train_dict[(u, i)] = r
```
